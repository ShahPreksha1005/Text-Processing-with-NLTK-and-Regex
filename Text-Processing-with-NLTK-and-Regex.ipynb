```python
paragraph = """
Social Media Marketing (SMM) is highly effective due to its capacity to connect
businesses with customers, facilitate dynamic interactions, and provide valuable
 customer data. Social media platforms, such as Facebook, Instagram, and YouTube,
 enable businesses to engage with target audiences in previously impossible ways.
  These platforms allow businesses to connect through various content forms,
  including videos, blogs, and direct communication, making the interaction not
  only diverse but also measurable. Through eWOM (electronic word-of-mouth),
  businesses can leverage free advertising opportunities where existing and
  potential customers share recommendations, driving consumer decisions and
  allowing the measurement of social equity.


Customer data is another critical resource provided by SMM. Advanced tools in
social media marketing can handle the 3Vs of big data (volume, variety, and
velocity), extracting valuable insights and enabling actionable market analysis.
 However, it is essential to consider that different demographics may have
 unequal access to social media, potentially excluding certain groups without
 online access. As social media platforms evolved, they transformed how
 businesses influence consumer behavior, promoting content to drive engagement
 and extracting demographic and personal information for targeted messaging.

An effective SMM action plan involves several steps: aligning SMM goals with
business objectives, understanding the target customer, conducting competitive
analysis, auditing current SMM efforts, creating a content delivery calendar,
producing high-quality content, and tracking performance to adjust strategies as
 needed. Tools like Hootsuite recommend setting clear objectives, learning about
  target demographics, and conducting regular audits to refine strategies.

SMM has distinct advantages over traditional marketing, particularly in Customer
 Relationship Management (CRM). SMM enables both customer-to-customer and
 firm-to-customer interactions, allowing businesses to track customer value
 through direct purchases and indirect referrals. The creation of sticky
 content, which engages customers at first glance and encourages sharing, is
 vital for driving growth through social media. Earned media, such as
 customer-created product reviews and recommendations, provides brand exposure
 without paid advertising. Viral marketing further amplifies this by triggering
 rapid word-of-mouth spread.

Customer segmentation is more refined in SMM, allowing businesses to target
specific demographics accurately. Tracking metrics like engagement, impressions,
 reach, share of voice, referrals, and conversions helps measure the
 effectiveness of SMM campaigns. It is crucial to align business goals with
 relevant metrics, using social media analytics tools to assess campaign
 performance against specific targets.

Despite its benefits, SMM has challenges, including the time and effort required
 to build a strong presence, the need for platform-specific expertise, and the
 unpredictability of platform algorithm changes. Additionally, public complaints
  can harm a company's reputation if not handled properly. Measuring the ROI of
  social media marketing can be complex due to the need to track multiple
  metrics and analyze data sets.

Overall, social media marketing is a powerful tool for building brands,
increasing sales, and driving website traffic. As social media usage grows
globally, businesses must navigate the competitive landscape to effectively
reach and engage user populations, maximizing the potential of SMM strategies.
"""
```

**(b) Write a program to print the number of total words and total unique words in the paragraph.**

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

# Tokenize the paragraph into words
words = word_tokenize(paragraph.lower())

# Remove stopwords and non-alphanumeric tokens
stop_words = set(stopwords.words('english'))
filtered_words = [word for word in words if word.isalnum() and word not in stop_words]

# Calculate total words and unique words
total_words = len(filtered_words)
unique_words = len(set(filtered_words))

print(f"Total words: {total_words}")
print(f"Total unique words: {unique_words}")
```

**(c) Find the frequency of all words and also display the most and least frequent word.**


```python
from collections import Counter

# Count the frequency of each word
word_freq = Counter(filtered_words)

# Find the most and least frequent words
most_common_word, most_common_count = word_freq.most_common(1)[0]
least_common_word, least_common_count = word_freq.most_common()[-1]

print(f"Word frequencies: {word_freq}")
print(f"Most frequent word: {most_common_word} (count: {most_common_count})")
print(f"Least frequent word: {least_common_word} (count: {least_common_count})")
```

**(d) Find the longest word in the paragraph.**

```python
# Find the longest word
longest_word = max(filtered_words, key=len)
print(f"Longest word: {longest_word}")
```

```python
import re

# Sample text
test_string = """"
1. Social media marketing is an 100py essential tool in the modern marketing grotto. The raven industry continues to evolve with new strategies. Social media allows brands to connect with their audience in a more personal way. A successful campaign relies on alphabetic strings of content that resonate with users. Consistent engagement with followers is crucial for maintaining a strong online presence. The bunnies are hopping around the new product launch, bringing excitement. The team's job is to create posts that are both informative and engaging. A great post should end with a compelling call to action. Some of the best campaigns have used clever tactics like repeating key phrases or words to emphasize important points.
b This approach can help to b attract and b retain bab attention, ensuring babble that the message is remembered. The power of social media lies in its ability to reach a vast audience quickly and efficiently.
"""
```

### **Exercise 2.1:**
#### **Exercise 2.1 typically involves using regular expressions to find specific patterns in text. Let's define a sample text and apply regular expressions to find patterns as per the exercise.**

```python
# Exercise 2.1
regex_alpha = r'\b[a-zA-Z]+\b'
regex_lowercase_b = r'\b[a-z]*b\b'
regex_ab = r'\bbab+\b'
```

#### **Exercise 2.2**
Exercise 2.2 typically involves more complex patterns or tasks related to regular expressions. Let's define another sample text and apply regular expressions accordingly.


```python
# Exercise 2.2
regex_repeated_words = r'\b(\w+)\s+\1\b'
regex_start_int_end_word = r'^\d.*\b\w+\b$'
regex_grotto_raven = r'\bgrotto\b.*\braven\b|\braven\b.*\bgrotto\b'
regex_first_word_register = r'^\s*([A-Za-z]+)'
```

```python
# Apply the regex patterns
matches_alpha = re.findall(regex_alpha, test_string)
matches_lowercase_b = re.findall(regex_lowercase_b, test_string)
matches_ab = re.findall(regex_ab, test_string)
matches_repeated_words = re.findall(regex_repeated_words, test_string)
matches_start_int_end_word = re.findall(regex_start_int_end_word, test_string)
matches_grotto_raven = re.findall(regex_grotto_raven, test_string)
matches_first_word_register = re.findall(regex_first_word_register, test_string)
```

###**Results:**

```python
print("Matches for alphabetic strings:", matches_alpha)
print("Matches for lowercase strings ending in 'b':", matches_lowercase_b)
print("Matches for 'a' surrounded by 'b':", matches_ab)
print("Matches for repeated words:", matches_repeated_words)
print("Matches for start with integer and end with word:", matches_start_int_end_word)
print("Matches for containing 'grotto' and 'raven':", matches_grotto_raven)
print("Matches for first word register:", matches_first_word_register)
```

```python
# Chapter 3 Text - The Power of Social Media Marketing
chapter_text = """
In today's digital age, social media marketing has become an essential tool for businesses looking to expand their reach and engage with their audience. This chapter delves into the various aspects of social media marketing, from creating engaging content to measuring success.

### Understanding Social Media Platforms

Before diving into strategies, it’s important to understand the unique characteristics of different social media platforms. Each platform has its own demographics, user behavior, and types of content that perform best.

1. **Facebook:** With over 2.8 billion monthly active users, Facebook remains the largest social media platform. It is ideal for businesses looking to build community and engage with customers through posts, comments, and messages. Visual content such as images and videos tend to perform well here.

2. **Instagram:** Known for its visual-centric approach, Instagram is perfect for brands with strong visual content. Stories, Reels, and IGTV provide various ways to showcase products, behind-the-scenes footage, and user-generated content.

3. **Twitter:** A platform for real-time updates and short messages, Twitter is effective for businesses that want to engage in conversations, provide customer service, and stay on top of trending topics.

4. **LinkedIn:** As a professional networking site, LinkedIn is best suited for B2B marketing. It's a great platform for sharing industry insights, company news, and connecting with other professionals.

5. **TikTok:** This rapidly growing platform is known for its short, engaging videos. Brands targeting a younger audience can leverage TikTok to create viral content and participate in trends.

### Crafting Engaging Content

Creating content that resonates with your audience is key to a successful social media strategy. Here are some tips:

- **Know Your Audience:** Understand who your audience is, what they like, and how they behave on social media. Use this information to tailor your content to their preferences.

- **Tell a Story:** People connect with stories. Use storytelling techniques to make your content more relatable and engaging. Whether it’s sharing customer testimonials, brand history, or behind-the-scenes looks, stories can make your brand more human.

- **Visual Appeal:** High-quality visuals are crucial on social media. Invest in good photography, design, and video production to create eye-catching content.

- **Consistency:** Maintain a consistent posting schedule and voice across all platforms. Consistency helps build brand recognition and trust with your audience.

### Engaging with Your Audience

Engagement is a two-way street. It’s not just about posting content, but also about interacting with your audience.

- **Respond to Comments and Messages:** Show your audience that you value their input by responding to comments and messages promptly. This interaction can foster a sense of community and loyalty.

- **Use Hashtags:** Hashtags can increase the visibility of your posts. Use relevant hashtags to reach a wider audience and join conversations.

- **Run Contests and Giveaways:** Contests and giveaways can boost engagement and attract new followers. Make sure the rules are clear and the prizes are appealing to your target audience.

- **Collaborate with Influencers:** Partnering with influencers can help you reach new audiences and add credibility to your brand. Choose influencers whose audience aligns with your target market.

### Measuring Success

To determine the effectiveness of your social media strategy, it’s important to track key metrics.

- **Reach and Impressions:** These metrics show how many people have seen your content. Reach refers to the number of unique users, while impressions count the total number of times your content is displayed.

- **Engagement Rate:** This metric measures the level of interaction your content receives, including likes, comments, shares, and clicks. A high engagement rate indicates that your content resonates with your audience.

- **Conversion Rate:** This metric tracks the number of users who take a desired action, such as making a purchase or signing up for a newsletter. It helps measure the ROI of your social media efforts.

- **Audience Growth:** Monitor the growth of your followers over time. A steady increase in followers indicates that your content is attracting new users.

### Conclusion

Social media marketing is a powerful tool that can help businesses connect with their audience, build brand loyalty, and drive sales. By understanding the unique characteristics of each platform, creating engaging content, interacting with your audience, and measuring your success, you can harness the full potential of social media marketing.
"""
```

```python
# Function to extract sections from the chapter text
def extract_section(section_name, text):
    """Extracts text under a specific section from the chapter."""
    start_marker = f"### {section_name}"
    end_marker = "###"  # End of the section
    start_index = text.find(start_marker)
    end_index = text.find(end_marker, start_index + len(start_marker))
    if start_index == -1 or end_index == -1:
        return ""
    return text[start_index:end_index].strip()
```

```python
# Extracting relevant sections
platforms_text = extract_section("Understanding Social Media Platforms", chapter_text)
content_text = extract_section("Crafting Engaging Content", chapter_text)
engagement_text = extract_section("Engaging with Your Audience", chapter_text)
success_text = extract_section("Measuring Success", chapter_text)
```

```python
# Helper function to count words and visualize them
def plot_most_common_words(text, title):
    """Count and plot the most common words in the given text."""
    words = re.findall(r'\b\w+\b', text.lower())  # Tokenize words
    word_counts = Counter(words)
    common_words = word_counts.most_common(10)  # Get top 10 common words

    # Plotting
    plt.figure(figsize=(10, 6))
    plt.bar([word[0] for word in common_words], [word[1] for word in common_words])
    plt.title(f"Top 10 Most Common Words in {title}")
    plt.xlabel("Words")
    plt.ylabel("Frequency")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
# Perform EDA on each section
plot_most_common_words(platforms_text, "Understanding Social Media Platforms")
plot_most_common_words(content_text, "Crafting Engaging Content")
plot_most_common_words(engagement_text, "Engaging with Your Audience")
plot_most_common_words(success_text, "Measuring Success")
```
